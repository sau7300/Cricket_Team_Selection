```python
import pandas as pd
import json
from pandas import json_normalize
with open(r'C:\Users\Hp\Downloads\data-analytics-project-for-beginners\t20_json_files\t20_json_files\t20_wc_match_results.json') as  f:
    data  = json.load(f)

    
df = json_normalize(data[0]['matchSummary'])
df
```

```python
df.shape

```

```python
df.rename({'scorecard':'matchid'} , inplace = True , axis =1)
df
```

```python
match_ids_dict = {}

for index, row in df.iterrows():
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1']
    match_ids_dict[key1] = row['matchid']
    match_ids_dict[key2] = row['matchid']
```

```python
df.to_csv('t20_csv_files/dim_match_summary.csv', index = False)
```

# batting

```python
with open(r'C:\Users\Hp\Downloads\data-analytics-project-for-beginners\t20_json_files\t20_json_files\t20_wc_batting_summary.json') as  file:
    data  = json.load(file)
    
all_records = []
for rec in data:
    all_records.extend(rec['battingSummary'])

df_batting = json_normalize(all_records)
df_batting
```

```python
for index, row in df_batting.iterrows():
    if len(row['dismissal']) > 0:
        df_batting.at[index, 'out/Not'] = 'out'
    else:
        df_batting.at[index, 'out/Not'] = 'Not_out'

df_batting.tail()
```

```python
df_batting.drop(columns=["dismissal"], inplace=True)
```

```python
df_batting['matchid'] = df_batting['match'].map(match_ids_dict)
df_batting
```

```python
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('', ''))
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('\xa0', ''))
df_batting.head()
```

```python
df_batting.to_csv('t20_csv_files/fact_bating_summary.csv', index = False)
```
